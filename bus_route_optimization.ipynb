```python
import pandas as pd
import numpy as np

# Assuming df1 and df2 are already defined as per your message.
# Here's the Haversine function to calculate distances.
def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance in kilometers between two points 
    on the earth (specified in decimal degrees)
    """
    # Convert decimal degrees to radians
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    
    # Haversine formula
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a)) 
    r = 6371 # Radius of earth in kilometers. Use 3956 for miles
    return c * r

# Extracting stop coordinates for easy lookup.
stop_coords = df1.set_index('stop_id')[['stop_lat', 'stop_lon']]

# Calculate distances for df2
df2['distance'] = np.nan  # Initialize the column with NaNs

for index, row in df2.iterrows():
    start_stop_id = int(row['start_stop'].split('_')[-1])  # Assuming the format is always correct
    end_stop_id = int(row['end_stop'].split('_')[-1])

    if start_stop_id in stop_coords.index and end_stop_id in stop_coords.index:
        # Get coordinates for the start and end stops
        start_lat, start_lon = stop_coords.loc[start_stop_id]
        end_lat, end_lon = stop_coords.loc[end_stop_id]
        
        # Calculate distance and update the dataframe
        df2.at[index, 'distance'] = haversine(start_lon, start_lat, end_lon, end_lat)

df2
```

```python
# Calculate distances from user's current location to each bus stop
df1['distance_to_user'] = df1.apply(lambda row: haversine(user_current_lon, user_current_lat, row['stop_lon'], row['stop_lat']), axis=1)
df1['distance_to_destination'] = df1.apply(lambda row: haversine(destination_lon, destination_lat, row['stop_lon'], row['stop_lat']), axis=1)

# Selecting columns for merging
columns_to_include_from_df3 = ['stop_id', 'stopid_line']

df3_unique = df3.drop_duplicates(subset=['stop_id'], keep='first')
# Filter nearest stops and merge with selected columns from df3
nearest_to_user = df1.nsmallest(3, 'distance_to_user')[['stop_id', 'stop_desc', 'stop_lat', 'stop_lon', 'distance_to_user']]
nearest_to_user = pd.merge(nearest_to_user, df3_unique[columns_to_include_from_df3], on='stop_id', how='inner')

nearest_to_destination = df1.nsmallest(3, 'distance_to_destination')[['stop_id', 'stop_desc', 'stop_lat', 'distance_to_destination']]
nearest_to_destination = pd.merge(nearest_to_destination, df3_unique[columns_to_include_from_df3], on='stop_id', how='inner')

nearest_to_user['stop_id'] = nearest_to_user['stop_id'].astype(str)

nearest_to_user['stopid_line'] = nearest_to_user['stopid_line'].astype(str)
df2['start_stop'] = df2['start_stop'].astype(str)
df2['end_stop'] = df2['end_stop'].astype(str)

# Try merging again with the correct columns
nearest_to_user = pd.merge(nearest_to_user, df2[['start_stop', 'interchange']], left_on='stopid_line', right_on='start_stop', how='left')
nearest_to_user.drop(columns=['start_stop'], inplace=True)

# If 'end_stop' could also be a match, you can attempt that merge too
nearest_to_destination = pd.merge(nearest_to_destination, df2[['end_stop', 'interchange']], left_on='stopid_line', right_on='end_stop', how='left')
nearest_to_destination.drop(columns=['end_stop'], inplace=True)
```

```python
start_stop_id = nearest_to_user['stopid_line'].iloc[0]
end_stop_id = nearest_to_destination['stopid_line'].iloc[0]
current_stop = start_stop_id
trip_route = [current_stop]
total_distance = 0
route_details = []

while current_stop != end_stop_id:
    next_stop_row = df2[df2['start_stop'] == current_stop]
    if next_stop_row.empty:
        print("No valid route found.")
        break
    next_stop = next_stop_row['end_stop'].values[0]
    interchange = next_stop_row['interchange'].values[0] 

    route_details.append({
        'from': current_stop,
        'to': next_stop,
        'distance': next_stop_row['distance'].values[0],
        'interchange': 'Exists' if interchange == 1 else 'None'
    })

    trip_route.append(next_stop)
    total_distance += next_stop_row['distance'].values[0]
    current_stop = next_stop

if current_stop == end_stop_id:
    route_output = " -> ".join([f"{detail['from']} -> {detail['to']}" for detail in route_details])
    print("Display Route : [" , route_output, "]")
    print("Total distance: {:.2f} kilometers".format(total_distance))
    print("Interchange Status: ", list(route_details[2].values())[3])
else:
    print("Could not complete the route to the final destination.")
```

```python
# Assuming df2 is already loaded with the appropriate data and includes necessary columns
# Columns assumed: ['interchange', 'trip_id', 'start_stop', 'end_stop', 'distance', 'Walking', 'Addprice']

start_stop_id = '1A_0_78'  # Example starting stop ID
end_stop_id = '210_0_77'  # Example ending stop ID
current_stop = start_stop_id
trip_route = [current_stop]
total_distance = 0
route_details = []

while current_stop != end_stop_id:
    # Find the next segment in the route that starts from the current stop
    next_segment = df2[df2['start_stop'] == current_stop]
    if next_segment.empty:
        print("No valid route found.")
        break
    
    # Assuming there might be multiple routes, pick the first or a specific condition
    next_stop = next_segment.iloc[0]['end_stop']
    distance = next_segment.iloc[0]['distance']
    interchange = next_segment.iloc[0]['interchange']  # Assuming 'interchange' denotes the type or presence of interchange
    # walking_time = next_segment.iloc[0]['Walking']
    
    # Append details for each leg of the trip
    route_details.append({
        'from': current_stop,
        'to': next_stop,
        'distance': distance,
        'interchange': 'Yes' if interchange else 'No',
        # 'walking_time': walking_time
    })
    
    trip_route.append(next_stop)
    total_distance += distance
    current_stop = next_stop

# Output the route and total distance
if current_stop == end_stop_id:
    print(f"Display Route from {start_stop_id} to {end_stop_id}:")
    for detail in route_details:
        print(f"From {detail['from']} to {detail['to']} ({detail['distance']} km, Interchange: {detail['interchange']})")
    print(f"Total distance: {total_distance:.2f} kilometers")
else:
    print("Could not complete the route to the final destination.")
```
